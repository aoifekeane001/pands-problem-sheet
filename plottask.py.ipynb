```python
### Problem 8:Write a program called plottask.py that displays:a histogram of a normal distribution of a 1000 values with a mean of 5 and standard deviation of 2, and a plot of the function  h(x)=x3 in the range [0, 10], on the one set of axes.
```

```python
### Solution 8:
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 values from a normal distribution with mean 5 and std dev 2
values = np.random.normal(5, 2, 1000)

# Plot a histogram of the values
plt.hist(values, bins=30, alpha=0.5, color='blue', edgecolor='black')

# Plot the function h(x) = x^3 in the range [0, 10]
x = np.linspace(0, 10, 100)
y = x ** 3
plt.plot(x, y, color='red')

# Set the title and axis labels
plt.title('Histogram of Normal Distribution and Plot of x^3')
plt.xlabel('Values')
plt.ylabel('Frequency / h(x)')

# Display the plot
plt.show()
```
